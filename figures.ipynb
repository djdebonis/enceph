```python
import scipy.stats as sstats
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = 20,10
plt.rcParams['font.size'] = 24
```

```python
n = sstats.norm(loc = 2, scale = 1)
n1 = sstats.norm(loc = 0, scale = 1)
```

```python
x = np.linspace(-5,5,100)
y = n.pdf(x)
z = n1.pdf(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.plot(x,z)
plt.tight_layout()
plt.title('Directional Selection')
plt.savefig('distributions.png')
plt.show()
```

```python
m = sstats.norm(loc = 0, scale = .7)
m1 = sstats.norm(loc = 0, scale = 1.7)
```

```python
xa = np.linspace(-5,5,100)
ya = m.pdf(xa)
za = m1.pdf(xa)
```

```python
plt.figure()
plt.plot(xa,ya)
plt.plot(xa,za)
plt.tight_layout()
plt.title('Stabilizing Selection')
plt.savefig('stablilizing_distribution.png')
plt.show()
```
